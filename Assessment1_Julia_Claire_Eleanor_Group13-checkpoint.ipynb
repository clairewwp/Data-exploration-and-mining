```python
import pandas as pd
import numpy as np
#df=pd.read_csv(D1.csv)
df = pd.read_csv('/Users/eleanor/OneDrive - Queensland University of Technology/SEM2/IFN509/Assessment1/D1.csv')

#data type correction
df['Date']=pd.to_datetime(df['Date'])#convert to datetime
df['Day'] = df['Day'].astype(str) #convert to string
df['Month']=df['Month'].astype(str)

#drop missing value in order to convert to int successfully 
na=df['Cloud9am'].dropna().astype(int)
df['Cloud9am']=na.astype(str)#then, conver to string
na=df['Cloud3pm'].dropna().astype(int)
df['Cloud3pm']=na.astype(str)

df['RainLastWeek']=df['RainLastWeek'].replace({'Yes':True,'No':False}).dropna()#drop null,transfer Yes/No to True/False
df['RainLastWeek']=df['RainLastWeek'].astype('boolean')#convert to boolean
df['RainToday']=df['RainToday'].replace({'Yes':True,'No':False}).dropna()
df['RainToday']=df['RainToday'].astype('boolean')
df['RainTomorrow']=df['RainTomorrow'].replace({'Yes':True,'No':False}).dropna()
df['RainTomorrow']=df['RainTomorrow'].astype('boolean')
df['ENSO']=df['ENSO'].replace({'Yes':True,'yes':True,'No':False,'.':np.nan}).dropna()#unify yes/Yes to True
df['ENSO']=df['ENSO'].astype('boolean')                                              #make '.' to null, and drop null

df['WindGustDir']=df['WindGustDir'].replace({'NE':'NY','WNW':'NWN'})#replacing the typo in wind direction data
df['WindDir9am']=df['WindDir9am'].replace({'NE':'NY','WNW':'NWN'})
df['WindDir3pm']=df['WindDir3pm'].replace({'NE':'NY','WNW':'NWN'})
```
